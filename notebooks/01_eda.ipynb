```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

# Configura√ß√µes
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette('husl')
pd.set_option('display.max_columns', None)

import warnings
warnings.filterwarnings('ignore')

print("‚úì Bibliotecas importadas!")
```

```python
# Ajuste o caminho para seu dataset
DATA_PATH = '../data/raw/creditcard.csv'

# Carregar
df = pd.read_csv(DATA_PATH)

print(f"Dataset carregado!")
print(f"   Linhas: {df.shape[0]:,}")
print(f"   Colunas: {df.shape[1]}")
print()

# Primeiras linhas
df.head()
```

```python
# Informa√ß√µes b√°sicas
print("Informa√ß√µes do Dataset:\n")
df.info()
```

```python
# Estat√≠sticas descritivas
df.describe()
```

```python
# Verificar valores ausentes
missing = df.isnull().sum()
if missing.sum() > 0:
    print("Valores ausentes:")
    print(missing[missing > 0])
else:
    print("‚úì Sem valores ausentes!")
```

```python
# Distribui√ß√£o de classes
fraud_counts = df['Class'].value_counts()
fraud_pct = df['Class'].value_counts(normalize=True) * 100

print("Distribui√ß√£o de Classes:\n")
print(f"N√£o-Fraude (0): {fraud_counts[0]:,} ({fraud_pct[0]:.2f}%)")
print(f"Fraude (1):     {fraud_counts[1]:,} ({fraud_pct[1]:.2f}%)")
print(f"\n‚ö†Ô∏è Ratio: 1 fraude para cada {fraud_counts[0]/fraud_counts[1]:.0f} transa√ß√µes normais")
```

```python
# Visualizar distribui√ß√£o
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Gr√°fico de barras
fraud_counts.plot(kind='bar', ax=axes[0], color=['#2ecc71', '#e74c3c'])
axes[0].set_title('Contagem de Transa√ß√µes', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Classe')
axes[0].set_ylabel('Quantidade')
axes[0].set_xticklabels(['Normal', 'Fraude'], rotation=0)

# Gr√°fico de pizza
colors = ['#2ecc71', '#e74c3c']
axes[1].pie(fraud_counts, labels=['Normal', 'Fraude'], autopct='%1.1f%%',
            colors=colors, startangle=90)
axes[1].set_title('Propor√ß√£o de Classes', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# Estat√≠sticas por classe
print("An√°lise de Valores (Amount):\n")
print("Transa√ß√µes Normais:")
print(df[df['Class']==0]['Amount'].describe())
print("\nTransa√ß√µes Fraudulentas:")
print(df[df['Class']==1]['Amount'].describe())
```

```python
# Visualizar distribui√ß√µes
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Distribui√ß√£o geral
df['Amount'].hist(bins=50, ax=axes[0, 0], edgecolor='black')
axes[0, 0].set_title('Distribui√ß√£o de Amount (Geral)', fontweight='bold')
axes[0, 0].set_xlabel('Valor')
axes[0, 0].set_ylabel('Frequ√™ncia')

# Por classe
df[df['Class']==0]['Amount'].hist(bins=50, ax=axes[0, 1], alpha=0.7, 
                                   label='Normal', color='#2ecc71')
df[df['Class']==1]['Amount'].hist(bins=50, ax=axes[0, 1], alpha=0.7, 
                                   label='Fraude', color='#e74c3c')
axes[0, 1].set_title('Amount por Classe', fontweight='bold')
axes[0, 1].legend()
axes[0, 1].set_xlabel('Valor')

# Boxplot
df.boxplot(column='Amount', by='Class', ax=axes[1, 0])
axes[1, 0].set_title('Compara√ß√£o de Amount', fontweight='bold')
axes[1, 0].set_xlabel('Classe')
plt.sca(axes[1, 0])
plt.xticks([1, 2], ['Normal', 'Fraude'])

# Log scale
df[df['Class']==0]['Amount'].apply(np.log1p).hist(bins=50, ax=axes[1, 1], 
                                                    alpha=0.7, label='Normal')
df[df['Class']==1]['Amount'].apply(np.log1p).hist(bins=50, ax=axes[1, 1], 
                                                    alpha=0.7, label='Fraude')
axes[1, 1].set_title('Amount (Log Scale)', fontweight='bold')
axes[1, 1].legend()

plt.tight_layout()
plt.show()
```

```python
# Se tiver coluna Time
if 'Time' in df.columns:
    # Converter para horas
    df['Hour'] = df['Time'] / 3600
    
    # Fraudes ao longo do tempo
    fig, ax = plt.subplots(figsize=(14, 6))
    
    df[df['Class']==0]['Hour'].hist(bins=48, ax=ax, alpha=0.5, 
                                     label='Normal', color='#2ecc71')
    df[df['Class']==1]['Hour'].hist(bins=48, ax=ax, alpha=0.5, 
                                     label='Fraude', color='#e74c3c')
    
    ax.set_title('Distribui√ß√£o Temporal de Transa√ß√µes', fontsize=14, fontweight='bold')
    ax.set_xlabel('Horas desde in√≠cio')
    ax.set_ylabel('Frequ√™ncia')
    ax.legend()
    plt.show()
```
